```python
import pandas as pd

pinleidf = pd.read_excel("附件1.xlsx")
chengbendf = pd.read_excel("附件3.xlsx")

chengbendf['销售日期'] = pd.to_datetime(chengbendf['日期'])
shai = (chengbendf['销售日期'] >= "2023-06-01") & (chengbendf['销售日期'] <= "2023-06-30")
chengbendf = chengbendf.loc[shai]
print("Done")
```

```python
df = chengbendf.merge(pinleidf,on="单品编码",how="left")
df
```

```python
danpindf = pd.read_excel("附件2.xlsx")
danpindf["销售日期"] = pd.to_datetime(danpindf["销售日期"])
print("Done")
```

```python
shai2 = (danpindf['销售日期'] >= "2023-06-01") & (danpindf['销售日期'] <= "2023-06-30")
danpindf = danpindf.loc[shai2]
```

```python
dftotal = danpindf.merge(df,on=["销售日期","单品编码"],how="left")
dftotal
```

```python
summary = dftotal.groupby(["分类名称","销售日期","是否打折销售", "单品名称","批发价格(元/千克)","销售单价(元/千克)"], as_index=False)["销量(千克)"].sum()
summary
```

```python
summary["加权平均加成率"] = (summary["销售单价(元/千克)"] / summary["批发价格(元/千克)"] - 1)*summary["销量(千克)"]
summary["加权平均成本"] = summary["批发价格(元/千克)"]*summary["销量(千克)"]
summary = summary.groupby(["分类名称","是否打折销售","销售日期"], as_index=False)[["销量(千克)","加权平均加成率","加权平均成本"]].sum()
summary["加权平均加成率"] = summary["加权平均加成率"] / summary["销量(千克)"]
summary["加权平均成本"] = summary["加权平均成本"] / summary["销量(千克)"]
summary
```

```python
with pd.ExcelWriter("销量加成关系.xlsx", engine="openpyxl") as writer:
    for pinlei,pinleiGroup in summary.groupby(["分类名称"]):
        sheetName1 = str(pinlei[0])
        for dazhe,dazheGroup in pinleiGroup.groupby(["是否打折销售"]):
            sheetName2 = sheetName1 + "_" + str(dazhe[0])
            dazheGroup.to_excel(writer, sheet_name=sheetName2, index=False)
            print(f"已完成：{sheetName2}")
print("任务完成")
```
